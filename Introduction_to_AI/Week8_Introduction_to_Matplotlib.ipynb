#Bar Graph

```python
Score = np.array([20, 35, 30, 35, 27])

plt.bar([0,1,2,3,4], Score)
plt.show()
```

```python
id1 = np.arange(1,10)
salary = np.arange(20,110,10)
plt.figure(figsize=(8,5)) # Setting the figure size
ax = plt.axes()
ax.set_facecolor("#ECF0F1") # Setting the background color by specifying the HEX Code
plt.bar(id1,salary,color = '#FFA726')
plt.xlabel(r'$Employee $ $ ID$')
plt.ylabel(r'$Salary$')
plt.show()
```

```python
#Plotting multiple sets of data
x1= [1,3,5,7]
x2=[2,4,6,8]
y1 = [7,8,15,10]
y2= [17,18,29,40]
plt.figure(figsize=(8,6))
ax = plt.axes()
ax.set_facecolor("white")
plt.bar(x1,y1,label = "First",color = '#42B300') # First set of data
plt.bar(x2,y2,label = "Second",color = '#94E413') # Second set of data
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title ('$Bar $ $ Chart$')
plt.legend()
plt.show()
```

Values and Label at the top of the graph


```python
barplot = plt.bar(id1, salary)
for bar in barplot:
  yval = bar.get_height()
  plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom') #va: vertical alignment y positional argument
    
plt.title(" Bar graph example")
plt.xlabel('Employees')
plt.ylabel("Salary")
```

Hide axis

```python
plt.figure(figsize=(7,7))
barplot = plt.bar(id1, salary)
for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom') #va: vertical alignment y positional argument
    plt.yticks([])
plt.title(" Bar graph example")
plt.xlabel('Employees')
plt.ylabel("Salary")
plt.show()
```

Grouped Bar Charts



```python
# Grouped Bar Chart
plt.figure(figsize=(7,9))
# set width of bar
barWidth = 0.25
# set height of bar
y1= np.array([17,18,29,40])
y2 =np.array([20,21,22,23])
y3 =np.array([5,9,11,12])
# Set position of bar on X axis
pos1 = np.arange(len(y1))
pos2 = [x + barWidth for x in pos1]
pos3 = [x + barWidth for x in pos2]
# Make the plot
plt.bar(pos1, y1, color='#FBC02D', width=barWidth, label='Batting')
plt.bar(pos2, y2, color='#F57F17', width=barWidth, label='Bowling')
plt.bar(pos3, y3, color='#E65100', width=barWidth, label='Fielding')
# Add xticks on the middle of the group bars
plt.xlabel('Player', fontweight='bold')
plt.ylabel('Value', fontweight='bold')
plt.xticks([i + barWidth for i in range(len(y1))], ['Shakib','Mashrafe','Mushfiqur','Ashraful'])

# Create legend & Show graphic
plt.legend()
plt.show()
```

Displaying values in Bar Charts

```python
# Displaying values in the stacked vertical bars using plt.text()
plt.style.use('seaborn-darkgrid')
x1= ['Shakib','Mashrafe','Mushfiqur','Ashraful']
y1= [17,18,29,40]
y2 = [20,21,22,23]
plt.figure(figsize=(5,7))
plt.bar(x1,y1,label = "Batting Average",width = 0.5,color = '#FF6F00')
plt.bar(x1,y2,label = "Bowling Average",width = 0.5 ,bottom = y1 , color = '#FFB300')
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title ('$Bar $ $ Chart$')
for x,y in zip(x1,y1):
    plt.text(x, y-10, '%d' % y, ha='center' , va= 'bottom')
for x,y,z in zip(x1,y2,y1):
    plt.text(x, y+z-10, '%d' % y, ha='center' , va= 'bottom')
plt.legend()
plt.show()
```

```python
# Displaying values in the stacked horizontal bars using plt.text()
plt.style.use('seaborn-darkgrid')
x1= ['Shakib','Mashrafe','Mushfiqur','Ashraful']
y1= [17,18,29,40]
y2 = [20,21,22,23]
plt.figure(figsize=(8,5))
plt.barh(x1,y1,label = "Batting Average",color = '#FF6F00')
plt.barh(x1,y2,label = "Bowling Average", left = y1 , color = '#FFB300')
plt.xlabel('$X$')
plt.ylabel('$Y$')
for x,y in zip(x1,y1):
    plt.text(y-10, x, '%d' % y, ha='center' , va= 'bottom')
for x,y,z in zip(x1,y2,y1):
    plt.text(y+z-10, x, '%d' % y, ha='center' , va= 'bottom')
plt.title ('$Bar $ $ Chart$')
plt.legend()
plt.show()
```

```python
# Shading Regions with fill_between() function
x = np.linspace(0, 10, 2000)
plt.figure(figsize=(10,6))
plt.plot(x,np.sin(x) , label = '$Sin(X)$')
plt.plot(x,np.cos(x) , label = '$cos(X)$')
plt.fill_between(x,0,np.sin(x))
plt.fill_between(x,0,np.cos(x))
plt.xlabel(r'$X$' , fontsize = 18)
plt.ylabel(r'$Y$' , fontsize = 18)
plt.title("$Sin(x) $ $ & $ $ Cos(x)$" ,fontsize = 14)
plt.legend(loc = 'lower left') # Legend will be placed at lower left position
plt.show()
```
